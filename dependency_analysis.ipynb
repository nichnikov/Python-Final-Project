```python
import pandas as pd
df = pd.read_csv("./clear_data.csv")
df
```

**1. коэффициенты корреляции между продажами товарных групп**

```python
# Сгруппируем продажи по товарным группам с помощью сводной таблицы
df_pt = pd.pivot_table(df[["Date", "Product_line", "Total"]], values=["Total"], index=["Date"], columns="Product_line")
df_pt
```

```python
# Построим матрицу корреляции между товарными категориями:
df_pt.corr()
```

Из матрицы корреляции видим, что товарные группы мало коррелируют между собой, самая высокая корреляция у товарных групп "Clothes and accessories" и "Fashion accessories", составляет **0.674**

**2. коэффициент корреляции между продажами Fashion accessories и Clothes and accessories для мужчин и для женщин**

```python
df.info()
```

**Мужчины**

```python
# отфильтруем продажи мужчинам по интересующим нас товарным группам
df_men = df[(df["Gender"] == "Male")&((df["Product_line"] == "Fashion accessories")|(df["Product_line"] == "Clothes and accessories"))]

# ссумируем продажи мужчинам по датам
df_men_pt = pd.pivot_table(df_men, values=["Total"], index=["Date"], columns="Product_line")

# посмотрим корреляцию между продажами товарных групп:
df_men_pt.corr()
```

Из таблицы видно, что корреляция между продажами "Fashion accessories" и "Clothes and accessories" для мужчин низкая

**Женщины**

```python
# отфильтруем продажи женщинам по интересующим нас товарным группам
df_women = df[(df["Gender"] == "Female")&((df["Product_line"] == "Fashion accessories")|(df["Product_line"] == "Clothes and accessories"))]

# ссумируем продажи мужчинам по датам
df_women_pt = pd.pivot_table(df_women, values=["Total"], index=["Date"], columns="Product_line")

# посмотрим корреляцию между продажами товарных групп:
df_women_pt.corr()
```

Из таблицы видно, что корреляция между продажами "Fashion accessories" и "Clothes and accessories" для женщин очень высокая. Можно уверенно сказать, что продажи этих двух товарных групп у женщин тесно взаимосвязана
